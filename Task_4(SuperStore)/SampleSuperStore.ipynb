```python
# Importing libraries 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
# Import datasets 
store_df = pd.read_csv('SampleSuperStore.csv')
# first five values
store_df.head()
```

last five values 

```python
store_df.tail()
```

```python
store_df.describe()
```

```python
store_df.isnull().any()
# no null values 
```

check for duplicated values 

```python
store_df.duplicated().sum()
```

Remove duplicated values 

```python
store_df.drop_duplicates(inplace=True)
```

Correlations between data

```python
sns.heatmap(store_df.corr(), square=True, cmap='Greens')
```

showing pairplot between every data 

```python
sns.pairplot(store_df)
```

#### Visualizing sales by state and cities 

Sales by state 

```python
pv = pd.pivot_table(store_df, index = store_df['State'], values='Sales', columns='Category')
plt.figure(figsize=(5,10))
sns.heatmap(pv, cmap="Greens")
```

Sales by city 

```python
pv = pd.pivot_table(store_df, index = store_df['City'], values='Sales', columns = 'Category')
plt.figure(figsize=(5,10))
sns.heatmap(pv,cmap="Greens")
```

Visualizing quantities by state and cities

Quantity by state

```python
pv = pd.pivot_table(store_df, index=store_df['State'], values='Quantity', columns = 'Category')
plt.figure(figsize=(5,10))
sns.heatmap(pv,cmap="Greens")
```

Quantity by city

```python
pv = pd.pivot_table(store_df, index=store_df['City'], values='Quantity', columns='Category')
plt.figure(figsize=(5,10))
sns.heatmap(pv, cmap="Greens")
```

#### Visualizing profits by sate and cities 

Profits by state

```python
pv = pd.pivot_table(store_df, index=store_df['State'], values='Profit', columns='Category')
plt.figure(figsize=(5,10))
sns.heatmap(pv, cmap="Greens")
```

Profile by city

```python
pv = pd.pivot_table(store_df, index=store_df['City'], values='Profit', columns='Category')
plt.figure(figsize=(5,10))
sns.heatmap(pv, cmap="Greens")
```

#### SubCategories

```python
plt.figure(figsize=(15,10))
store_df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

#### Plotting profits per sub-categories 

```python
plt.figure(figsize=(17,8))
sns.barplot(store_df['Sub-Category'], store_df['Profit'])
```
